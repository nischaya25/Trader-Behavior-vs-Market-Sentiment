```python
%%writefile app.py
import streamlit as st
import pandas as pd
import plotly.express as px

# -----------------------
# PAGE SETTINGS
# -----------------------
st.set_page_config(page_title="Trader Dashboard", layout="wide")

# -----------------------
# BACKGROUND THEME (CSS)
# -----------------------
page_bg = """
<style>
.stApp {
    background: linear-gradient(to right, #141e30, #243b55);
    color: white;
}

h1, h2, h3, h4 {
    color: #ffffff;
}

[data-testid="stSidebar"] {
    background-color: #1f2c3a;
}
</style>
"""

st.markdown(page_bg, unsafe_allow_html=True)

# -----------------------
# LOAD DATA
# -----------------------
df = pd.read_csv("/content/processed_trader_data (7).csv")
df.columns = df.columns.str.strip()

st.title("üìä Trader Behavior Dashboard")

# -----------------------
# SIDEBAR
# -----------------------
st.sidebar.header("Filters")

sentiment = st.sidebar.selectbox(
    "Select Sentiment",
    df['classification'].dropna().unique()
)

chart_type = st.sidebar.selectbox(
    "Select Chart Type",
    ["Bar", "Line", "Pie"]
)

x_axis = st.sidebar.selectbox(
    "X-axis",
    df.columns
)

y_axis = st.sidebar.selectbox(
    "Y-axis",
    df.select_dtypes(include='number').columns
)

# -----------------------
# FILTER DATA
# -----------------------
filtered = df[df['classification'] == sentiment]

# -----------------------
# METRICS
# -----------------------
st.subheader("Key Metrics")

avg_pnl = filtered['Closed PnL'].mean()
win_rate = (filtered['Closed PnL'] > 0).mean()

col1, col2 = st.columns(2)

col1.metric("Average PnL", round(avg_pnl, 2))
col2.metric("Win Rate", f"{round(win_rate*100,2)}%")

# -----------------------
# CHART DATA
# -----------------------
chart_data = (
    filtered.groupby(x_axis)[y_axis]
    .mean()
    .reset_index()
)

# -----------------------
# CHART
# -----------------------
st.subheader("Visualization")

if chart_type == "Bar":
    fig = px.bar(chart_data, x=x_axis, y=y_axis)

elif chart_type == "Line":
    fig = px.line(chart_data, x=x_axis, y=y_axis, markers=True)

elif chart_type == "Pie":
    fig = px.pie(chart_data, names=x_axis, values=y_axis)

st.plotly_chart(fig, use_container_width=True)

# -----------------------
# TABLE
# -----------------------
st.subheader("Filtered Data")
st.dataframe(filtered)
```
